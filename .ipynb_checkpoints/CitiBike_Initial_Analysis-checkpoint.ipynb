```python
#bikedata = feather.read_dataframe('../../Data/CitiBike_Data/bikedata.feather')
bikedata = pd.read_parquet("./data/CitiBike_Data/bikedata_201603.parquet")
print(bikedata.columns)
print(len(bikedata))
```

```python
# How many bikes will leave a particular station at a particular time?
# First let's start with a single station. In our case, let's start with E 40 St & 5 Ave
# dfFocusStation = bikedata[(bikedata['start station name'] == "8 Ave & W 31 St") 
#                           & (bikedata['usertype'] == 'Subscriber')]
dfFocusStation = bikedata[(bikedata['start station name'] == "8 Ave & W 31 St")]
dfGroupBy = dfFocusStation.groupby(by=['dtstartdatehour'])

# Number of departures per hour
departures = dfGroupBy.bikeid.count().reset_index()

# startdatehour - to join with hourly weather data
departures['startdatehour'] = departures.dtstartdatehour.apply(lambda x:x.strftime('%Y-%m-%d %H:%M:%S'))
# startdate - to join with public holiday data
departures['startdate'] = departures.dtstartdatehour.dt.date.apply(lambda x:x.strftime('%Y-%m-%d'))

# Male to female ratio
sum_male = dfGroupBy.male.sum().reset_index().male
sum_female = dfGroupBy.female.sum().reset_index().female
departures['male_to_female_ratio'] = (sum_male).astype(float) / (sum_male + sum_female)

print departures.head()
```

```python
departures['startdatehour'] = departures.dtstartdatehour.dt.hour
# plt.boxplot([(final_df.bikeid), (final_df.dtstartdatehour.dt.hour)])
departures.boxplot(column='bikeid', by='startdatehour')
```

```python
# Takes a lot of time
startHour = bikedata.starttime.str.split().str[1].str.split(':').str[0].astype('int')
endHour = bikedata.stoptime.str.split().str[1].str.split(':').str[0].astype('int')
```

```python
# Takes a lot of time
dfMorning = bikedata[endHour <= 10]
dfEvening = bikedata[endHour >= 16]

dfMorningStart = dfMorning.groupby(by=['start station id'])['bikeid'].count().sort_values(ascending=False)
dfMorningEnd = dfMorning.groupby(by=['end station id'])['bikeid'].count().sort_values(ascending=False)

dfEveningStart = dfEvening.groupby(by=['start station id'])['bikeid'].count().sort_values(ascending=False)
dfEveningEnd = dfEvening.groupby(by=['end station id'])['bikeid'].count().sort_values(ascending=False)
```

```python
# Master data for Station coordinates
coord1 = bikedata.drop_duplicates(subset='end station id')[['end station id', 'end station latitude', 'end station longitude']]
coord1.columns=['stationid', 'station_latitude', 'station_longitude']

coord2 = bikedata.drop_duplicates(subset='start station id')[['start station id', 'start station latitude', 'start station longitude']]
coord2.columns=['stationid', 'station_latitude', 'station_longitude']

coord3 = pd.concat([coord1, coord2])
station_coordinates = coord3.drop_duplicates(subset='stationid')
```

```python
# Data preparation for morning data

# Join morning start and morning end data. 
# We will plot staton wise data in the map. 
morning_df = pd.merge(dfMorningStart.reset_index(), dfMorningEnd.reset_index(), 
         left_on='start station id', right_on = 'end station id', how='outer')

morning_final_df = pd.DataFrame(columns=['stationid', 'net_bikeid'])
morning_final_df['net_bikeid'] = morning_df.bikeid_x - morning_df.bikeid_y

if morning_df['end station id'].isnull().sum() > morning_df['start station id'].isnull().sum():
    morning_final_df['stationid'] = morning_df['end station id']
else:
    morning_final_df['stationid'] = morning_df['start station id']
morning_final_df['color'] = morning_final_df.net_bikeid.apply(lambda x:'green' if x > 0 else 'red')

morning_final_df = pd.merge(morning_final_df, station_coordinates, on='stationid', how='left')
```

```python
evening_df = pd.merge(dfEveningStart.reset_index(), dfEveningEnd.reset_index(), 
         left_on='start station id', right_on = 'end station id', how='outer')

evening_final_df = pd.DataFrame(columns=['stationid', 'net_bikeid'])
evening_final_df['net_bikeid'] = evening_df.bikeid_x - evening_df.bikeid_y

if evening_df['end station id'].isnull().sum() > evening_df['start station id'].isnull().sum():
    evening_final_df['stationid'] = evening_df['end station id']
else:
    evening_final_df['stationid'] = evening_df['start station id']
evening_final_df['color'] = evening_final_df.net_bikeid.apply(lambda x:'green' if x > 0 else 'red')

evening_final_df = pd.merge(evening_final_df, station_coordinates, on='stationid', how='left')
```

```python
# Plotting just the morning traffic
nyMapMorning = folium.Map(location=[40.7831, -73.9712], zoom_start=12)

for index, row in morning_final_df.iterrows():
    folium.CircleMarker([row.get(3), row.get(4)], radius=20, color=row.get(2)).add_to(nyMapMorning)

nyMapMorning
```

```python
# Plotting just the evening traffic
nyMapEvening = folium.Map(location=[40.7831, -73.9712], zoom_start=12)

for index, row in evening_final_df.iterrows():
    folium.CircleMarker([row.get(3), row.get(4)], radius=20, color=row.get(2)).add_to(nyMapEvening)

nyMapEvening
```
